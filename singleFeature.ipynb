```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_train = np.array([200,500,400,300])
y_train = np.array([130,270,240,200])
y_hat = np.zeros(x_train.shape[0]);
```

```python
plt.scatter(x_train,y_train)
plt.show()
```

```python
def cost_function(x,y,w,b):
    m = x.shape[0]
    cost_sum = 0
    for i in range(m):
        f_wb = w*x[i] + b
        y_hat[i] = f_wb
        cost_sum+= (f_wb - y[i])**2
    cost_sum = cost_sum/(2*m)
    return cost_sum
    
```

```python
print(cost_function(x_train,y_train,0.6,0))
print(y_hat)
```

```python

plt.scatter(x_train,y_train)
plt.plot(x_train,y_hat)
```

```python
costs = np.zeros(20)
w_range = np.zeros(20)
def plt_prediction_cost(x,y,w,b):
    # plt.scatter(x,y)
    # plt.plot(x,y_hat)
    
    m = x.shape[0]
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10, 5))
    plt.subplots_adjust(wspace=0.5, hspace=1)
    ax1.scatter(x,y)
    ax1.set_title("House Price Predection")
    ax1.set_xlabel("sq feet")
    ax1.set_ylabel("Price")
    ax1.plot(x,y_hat)
    
    
   
    i = 0
    j = 0
    while(i <1):
        cost = cost_function(x,y,i,b)
        costs[j] = cost
        w_range[j] = i
        i+=0.05
        j+=1
        
    ax2.plot(w_range,costs)
    ax2.set_title("Cost vs w(b fixed at 0)")
    ax2.set_xlabel("w")
    ax2.set_ylabel("w")
    plt.tight_layout()
plt_prediction_cost(x_train,y_train,0.6,0)
    
```

```python
def compute_gradient(x,y,w,b):
    m = x.shape[0]
    dj_w = 0
    for i in range(m):
        f_wb = w*x[i] + b
        dj_w += (f_wb-y[i])*x[i]
    dj_w = dj_w/m

    return dj_w

def gradient_descent(x,y,w,iterations,alpha = 0.00001):
    temp_w = w
    d = []
    j = []
    
    for i in range(iterations):
        dj_w = compute_gradient(x,y,temp_w,0)
        
        temp_w = temp_w - alpha* dj_w
        d.append(temp_w)
        j.append(i)
    return temp_w,d,j
            

```

```python
w,d,j = gradient_descent(x_train,y_train,0.1,1000)
print(w)
```
